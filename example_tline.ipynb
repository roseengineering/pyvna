```python
%matplotlib inline
import pyvna as vna
vna.init(port='/dev/ttyUSB0', name='tiny')
```

```python
vna.temperature()
```

```python
vna.version()
```

```python
cal = vna.create(start=35e6, stop=1300e6)
```

```python
gm = vna.cal_open(cal, average=1)
vna.polar(gm)
vna.bode(gm, log=True)
```

```python
gm = vna.cal_short(cal, average=1)
vna.polar(gm)
vna.bode(gm, log=True)
```

```python
gm = vna.cal_load(cal, average=1)
vna.polar(gm)
vna.bode(gm, log=True)
```

```python
orig = cal
```

```python
# cal = vna.create(start=25e6, stop=600e6, resample=orig)
gm = vna.return_loss(cal, average=1)
vna.polar(gm)
vna.cartesian(gm)
y = vna.to_admittance(gm)
z = vna.to_impedance(gm)
xp = vna.to_parallel(gm)
vna.cartesian(z)
vna.cartesian(y)
vna.cartesian(xp)
vna.plot(z)
vna.plot(vna.to_inductance(z))
vna.plot(vna.to_capacitance(z))
vna.plot(y)
vna.plot(xp)
vna.plot(vna.to_inductance(xp))
vna.plot(vna.to_capacitance(xp))
vna.bode(gm)
vna.bode(gm, log=True)
```
