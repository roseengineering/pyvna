```python
%matplotlib inline
import pyvna as vna
vna.init(port='/dev/ttyUSB1', name='zero')
```

```python
cal = vna.create(start=400e3, points=200)
```

```python
gm = vna.cal_open(cal)
vna.polar(gm)
```

```python
gm = vna.cal_short(cal)
vna.polar(gm)
```

```python
gm = vna.cal_load(cal)
vna.polar(gm)
```

```python
# ferrite bead (fb43)
gm = vna.return_loss(cal)
vna.polar(gm)
vna.cartesian(gm)
y = vna.to_admittance(gm)
z = vna.to_impedance(gm)
xp = vna.to_parallel(gm)
vna.cartesian(z)
vna.cartesian(y)
vna.cartesian(xp)
vna.plot(z)
vna.plot(vna.to_inductance(z))
vna.plot(vna.to_capacitance(z))
vna.plot(y)
vna.plot(xp)
vna.plot(vna.to_inductance(xp))
vna.plot(vna.to_capacitance(xp))
vna.bode(gm)
```
